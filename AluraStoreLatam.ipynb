### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
t = {1:tienda1,
     2:tienda2,
     3:tienda3,
     4:tienda4}

tiendas = [t[i+1]['Precio'].sum() for i in range(len(t))]
for i in range(len(t)):
    print("Tienda", i+1, "facturación", tiendas[i])
```

```python
y = tiendas
l = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
plt.pie(y, labels = l, autopct = '%1.1f%%')
plt.legend(title = "Tiendas:")
plt.show()

x = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
y = tiendas

plt.bar(x,y)
plt.show()

xpoints = tiendas
ypoints = l

plt.plot(ypoints, xpoints, linestyle = 'dotted')
plt.show()
```

# 2. Ventas por categoría

```python
ventas = [t[i+1]['Categoría del Producto'].value_counts() for i in range(len(t))]
for i in range(len(t)):
    print("Tienda", i+1, "ventas", ventas[i])
```

# 3. Calificación promedio de la tienda


```python
calificacion = [t[i+1]['Calificación'].mean() for i in range(len(t))]
for i in range(len(t)):
    print("Tienda", i+1, "Calificación", calificacion[i])
```

```python
y = calificacion
l = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
plt.pie(y, labels = l, autopct = '%1.1f%%')
plt.legend(title = "Tiendas:")
plt.show()

x = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
y = calificacion

plt.bar(x,y)
plt.show()

xpoints = calificacion
ypoints = l

plt.plot(ypoints, xpoints, linestyle = 'dotted')
plt.show()
```

# 4. Productos más y menos vendidos

```python
mas = [t[i+1]['Producto'].value_counts().head(1) for i in range(len(t))]
menos = [t[i+1]['Producto'].value_counts().tail(1) for i in range(len(t))]
for i in range(len(t)):
    print(f"\n Tienda  {i+1} \n Producto mas vendido : \n {mas[i]}")
    print(f"Producto menos vendido : \n {menos[i]}")
```

# 5. Envío promedio por tienda

```python
envio = [t[i+1]['Costo de envío'].mean() for i in range(len(t))]
for i in range(len(t)):
    print("Tienda", i+1, "Envio", envio[i])

```

```python
y = envio
l = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
plt.pie(y, labels = l, autopct = '%1.1f%%')
plt.legend(title = "Tiendas:")
plt.show()

x = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
y = envio

plt.bar(x,y)
plt.show()

xpoints = envio
ypoints = l

plt.plot(ypoints, xpoints, linestyle = 'dotted')
plt.show()
```

Informe de Evaluación para Cierre de Tienda

Introducción
El propósito de este informe es apoyar la toma de decisiones del propietario respecto al cierre de una de sus tiendas, con base en criterios objetivos que incluyen facturación, calificación de los clientes y costos de envío. El análisis comparativo busca identificar cuál de las tiendas presenta un desempeño inferior en relación con las demás, considerando sus fortalezas y debilidades. Esta recomendación se elabora desde una perspectiva estratégica, buscando optimizar la eficiencia operativa y mantener la calidad del servicio al cliente.

Análisis Comparativo
1. Facturación:

Tienda 1: $1,150,880,400.00

Tienda 2: $1,116,343,500.00

Tienda 3: $1,098,019,600.00

Tienda 4: $1,038,375,700.00

La Tienda 4 es la que menor facturación presenta, con una diferencia significativa respecto a la Tienda 1, que lidera en ventas.

2. Calificación por clientes (de 0 a 5):

Tienda 1: 3.98

Tienda 2: 4.04

Tienda 3: 4.05

Tienda 4: 3.99

En cuanto a satisfacción del cliente, Tienda 4 también se encuentra por debajo de Tiendas 2 y 3, aunque apenas ligeramente por encima de Tienda 1.

3. Costo promedio de envío:

Tienda 1: $26,018.61

Tienda 2: $25,216.24

Tienda 3: $24,805.68

Tienda 4: $23,459.46

Tienda 4 destaca en este rubro por tener el costo de envío más bajo, lo cual es una fortaleza operativa.

Fortalezas y Debilidades por Tienda
Tienda	Fortalezas	Debilidades
1	Alta facturación	Bajo puntaje de calificación y alto costo de envío
2	Buen equilibrio entre calificación y facturación	Costo de envío elevado
3	Mejor calificación por clientes	Facturación inferior a Tienda 1 y 2
4	Menor costo de envío	Menor facturación general y calificación moderada

Conclusión y Recomendación
Tras analizar los datos disponibles, se concluye que la Tienda 4 es la candidata más adecuada para su cierre. Aunque presenta el costo de envío más bajo (lo cual representa eficiencia logística), sus niveles de facturación son los más bajos entre las cuatro tiendas. Además, su calificación de satisfacción del cliente no es suficientemente alta como para compensar su bajo rendimiento financiero. Las otras tiendas, aunque con algunas debilidades, tienen mejor desempeño en los indicadores clave y mayor potencial de mejora o consolidación.

Por lo tanto, se recomienda cerrar la Tienda 4, permitiendo que los recursos operativos y humanos se redistribuyan hacia las demás tiendas con mayor rentabilidad y mejor percepción por parte de los clientes.
